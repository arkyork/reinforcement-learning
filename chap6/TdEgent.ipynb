```python
from GridWorld import GridWorld
from TdAgent import TdAgent
env = GridWorld()
agent = TdAgent()
```

```python
episodes = 1000

for episode in range(episodes):
    # エピソードごとにリセット
    state = env.reset()

    while True:
        action = agent.get_action(state)
        next_state, reward, goal_done = env.step(action)

        agent.eval(state, reward, next_state, goal_done)

        if goal_done:
            break
        state = next_state
```

```python
env.render_value_function(agent.V)
```

```python
from GridWorld import argmax
pi = env.greedy_policy(agent.V,env,gamma=0.9)

action_command = {
    0: "↑",   # up
    1: "↓",   # down
    2: "←",   # left
    3: "→"    # right
}

count = 0
for pos,p in pi.items():
    count += 1
    print(action_command[argmax(p)],end=" ")
    if count % 4 == 0:
        print()
```
