```python
from SarasaOn import SarasaOnAgent
from GridWorld import GridWorld

env = GridWorld()
agent = SarasaOnAgent()
```

```python
episodes = 10000

for episode in range(episodes):
    state = env.reset()
    agent.reset()

    while True:
        action = agent.get_action(state)
        next_state,reward,goal_done = env.step(action)
        
        agent.update(state, action, reward, goal_done)

        if goal_done:
            agent.update(next_state, None, None, None)
            break
        state = next_state
```
