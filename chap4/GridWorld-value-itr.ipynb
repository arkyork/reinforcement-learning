```python
from GridWorld import GridWorld
from collections import defaultdict
from GridWorld import argmax
```

```python
V = defaultdict(lambda : 0)
env = GridWorld()

gamma = 0.9

V = env.value_itr(V=V,env=env,gamma=gamma)
```

```python
env.render_value_function(V)
```

```python
pi = env.greedy_policy(V,env,gamma)
```

```python
pi
```

```python
action_command = {
    0: "↑",   # up
    1: "↓",   # down
    2: "←",   # left
    3: "→"    # right
}

count = 0
for pos,p in pi.items():
    count += 1
    print(action_command[argmax(p)],end=" ")
    if count % 4 == 0:
        print()
```
