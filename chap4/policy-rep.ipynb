```python

# 価値反復
# V-> 価値関数
V = {"L1":0,"L2":0}

new_V = V.copy()
# 閾値の設定を行う

count = 0

while True:
    new_V["L1"] = 0.5 * ( -1 + 0.9 * V["L1"]) + 0.5 * ( 1 + 0.9 * V["L2"])
    new_V["L2"] = 0.5 * ( 0 + 0.9 * V["L1"]) + 0.5 * ( -1 + 0.9 * V["L2"])

    delta_V_1 = abs(new_V["L1"]-V["L1"])
    delta_V_2 = abs(new_V["L2"]-V["L2"])

    # delta max関数で差が大きいほうを選択

    delta = max(delta_V_1,delta_V_2)

    count += 1
    

    if delta < 0.0001:
        print(V)
        print(count)
        break

    V = new_V.copy()



    print(V)
```

## Vだけで実装
new_Vを使わずに動的計画法で実装

```python

# 価値反復
# V-> 価値関数
V = {"L1":0,"L2":0}

# 閾値の設定を行う

count = 0

while True:
    # V1の更新
    V_1_temp = 0.5 * ( -1 + 0.9 * V["L1"]) + 0.5 * ( 1 + 0.9 * V["L2"])
    delta_V_1 = abs(V["L1"] - V_1_temp)
    V["L1"] = V_1_temp

    # V2の更新
    V_2_temp = 0.5 * ( 0 + 0.9 * V["L1"]) + 0.5 * ( -1 + 0.9 * V["L2"])
    delta_V_2 = abs(V["L2"] - V_2_temp)
    V["L2"] = V_2_temp

    # delta max関数で差が大きいほうを選択
    delta = max(delta_V_1,delta_V_2)
    count += 1
    

    if delta < 0.0001:
        print(V)
        print(count)
        break

    print(V)
```
